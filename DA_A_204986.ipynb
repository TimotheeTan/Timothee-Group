```python
problem = "problem_1"
password= "a"
positive = 0
negative = 0
zeroes= 0
while True:
    guess = input("Please input your guess:")
    if guess == password:
        break
    elif int(guess) > 0: 
        positive += 1
        continue
    elif int(guess) < 0:
        negative += 1
        continue
    elif int(guess) == 0:
        zeroes += 1
        continue   
print ("Positive:",positive)
print ("Negative:",negative)
print ("Zeroes", zeroes)
```
