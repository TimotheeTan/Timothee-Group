```python
legs = {
     ("upd","admu"):{
         "travel_time_mins":10
     },
     ("admu","dlsu"):{
         "travel_time_mins":35
     },
     ("dlsu","upd"):{
         "travel_time_mins":55
     }
}

def eta(legs,source,destination):
    
    total_time = 0
    old_destination = ""
    keys = list(legs.keys())
    
    sources_list = [a[0] for a in keys]
    destinations_list = [a[1] for a in keys]
    while old_destination != destination:
        if (source,destination) in keys:
            old_destination = destination 

        elif (source,destination) not in keys:
             for i in range(len(legs)):
                if sources_list[i] == source:
                    old_destination = destinations_list[i]
                    break
                    
        travel_time = legs[(source,old_destination)]["travel_time_mins"]
        total_time += travel_time
        source = old_destination
    return total_time
eta (legs, "admu", "dlsu")
    
```
